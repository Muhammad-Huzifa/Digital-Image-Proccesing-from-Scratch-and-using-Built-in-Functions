## Mixing of Color wwith Different Proportion

```python
img = np.zeros((10,10,3) , dtype = np.uint8)
img[:,:,0] = 255
plt.imshow(img)
```

```python
red = np.array([255,0,0])
white = np.array([255,255,255])
proportion = [(1,0)  , (0.75 , 0.25)  , (.5 , 0.5) , (0.25 , 0.75) , (0,1)]
fig, ax = plt.subplots(1, len(proportion), figsize=(12, 12));

for i , (b,w) in enumerate(proportion):
    color = b*red + w * white
    img[:,:,:] = color
    ax[i].imshow(img)
    # ax[i].set_xlim(0,1)
    # ax[i].set_xlim(0,1)
    ax[i].set_title(f'({w,b})')
    ax[i].axis('off')
    
```

```python
img = cv.imread("D:\DIP Book\DIP3E_CH06_Original_Images\DIP3E_Original_Images_CH06\Fig0620(a)(picker_phantom).tif",0)
plt.imshow(img, 'gray')
```

```python
pseudo_img = np.zeros((img.shape[0] , img.shape[1] , 3 ) , dtype = np.uint8)
```

```python
pseudo_img[:,:,0] = img
pseudo_img[:,:,1] = img
pseudo_img[:,:,2] = img
plt.imshow(pseudo_img)
```

```python
for i in range(pseudo_img.shape[0]):
    for j in range(pseudo_img.shape[1]):
        if img[i,j] > 100 and img[i,j] <= 120:
            pseudo_img[i,j] = [255,60,30]
        elif img[i,j] >120  and img[i,j] <= 150:
            pseudo_img[i,j] = [255,255,255]
        else:
            pass

plt.imshow(pseudo_img)
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if (img[i,j]>100 and img[i,j]<120):
            pseudo_img[i,j] = [255,80,80]
        if (img[i,j]>120 and img[i,j]<130):
            pseudo_img[i,j] = [255,255,255] 
        if (img[i,j]>80 and img[i,j]<100):
            pseudo_img[i,j] = [255,255,0]
        if (img[i,j]>60 and img[i,j]<80):
            pseudo_img[i,j] = [0,255,0]
        if (img[i,j]>=50 and img[i,j]<60):
            pseudo_img[i,j] = [0,0,255]
        # if (img[i,j]>= 40 and img[i,j] <50):
        #     pseudo_img[i,j] = [0,0,0]

fig, ax = plt.subplots(1,2, figsize=(10,10));

ax[0].imshow(img, 'gray');
ax[0].set_title("Original Image");
ax[0].axis('off');

ax[1].imshow(pseudo_img, 'gray');
ax[1].set_title("Pseudocolored Image");
ax[1].axis('off');
```
