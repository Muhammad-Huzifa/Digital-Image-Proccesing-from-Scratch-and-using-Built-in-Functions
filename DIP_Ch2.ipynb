### Image Quantization

```python
img = cv.imread("C:\\Users\PMLS\Desktop\Images For DIP\DIP3E_CH02_Original_Images\DIP3E_Original_Images_CH02\Fig0221(a)(ctskull-256).tif",0)
```

```python
img_7bit  = (img // 2).astype('uint8') * 2
img_6bit  = (img // 4).astype('uint8') * 4
img_5bit  = (img // 8).astype('uint8') * 8
img_4bit  = (img // 16).astype('uint8') * 16
img_3bit  = (img // 32).astype('uint8') * 32
img_2bit  = (img // 64).astype('uint8') * 64
img_1bit  = (img // 128).astype('uint8') * 128
```

```python
fig , ax = plt.subplots(3,3 , figsize=(15,10))
ax[0,0].imshow(img_7bit , cmap = 'gray')
ax[0,0].axis('off')
ax[0,0].set_title('7bit')
ax[0,1].imshow(img_6bit , cmap= 'gray')
ax[0,1].axis('off')
ax[0,2].imshow(img_5bit , cmap= 'gray')
ax[0,2].axis('off')
ax[1,0].imshow(img_4bit , cmap= 'gray')
ax[1,0].axis('off')
ax[1,1].imshow(img_3bit , cmap= 'gray')
ax[1,1].axis('off')
ax[1,2].imshow(img_2bit , cmap= 'gray')
ax[1,2].axis('off')

ax[2,0].imshow(img_1bit , cmap= 'gray')
ax[2,0].axis('off')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an empty 2D NumPy array (e.g., 255x255 image)
img_height = 255
img_width = 255

# Create 17 distinct grayscale levels from black (0) to white (255)
levels = np.linspace(0, 255, 17, dtype=np.uint8)

# Repeat the levels to cover the width of the image
img = np.repeat(levels, img_width // 17)  # Repeat each level over a portion of the width

# Trim to exactly fit the width of the image if necessary (cut any extra pixels)
img = img[:img_width]

# Expand the array vertically to apply the same levels to all rows
img = np.tile(img, (img_height, 1))

# Display the image
plt.imshow(img, cmap='gray')
plt.title('17 Distinct Grayscale Levels (Left Black, Right White)')
plt.show()
```
